```python
import os
```

```python
folder = "/content/drive/MyDrive/pages"
```

```python
documents = {}
for file in os.listdir(folder):
  if file.endswith(".html"):
    with open(os.path.join(folder,file),"r",encoding="utf-8") as f:
      documents[file] = f.read().lower()
```

```python
print("Inverted files:",list(documents.keys()))
```

```python
query = input("enter your search query").lower()
result = []
for filename , content in documents.items():
  if query in content:
    result.append(filename)
```

```python
if result:
    print("\nüîç Search Results:")
    for r in result:
        print("‚Üí", r)
else:
    print("\nNo pages found matching your query.")
```
