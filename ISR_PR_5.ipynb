```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/content/drive/MyDrive/flower.jpeg"
img = cv2.imread(image_path)

if img is None:
  print("Error:Image not found")
  exit()
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
colors = ('r','g','b')
plt.figure(figsize=(8,5))
for i,col in enumerate(colors):
  hist = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist,color=col,linewidth=1.5)

plt.title("COLOR Flower")
plt.show()
```

```python
mean_colors = cv2.mean(img)[:3]
print("Mean Color : ")
print(f"Red:{mean_colors[2]:.2f}")
print(f"green:{mean_colors[1]:.2f}")
print(f"blue:{mean_colors[0]:.2f}")
```
